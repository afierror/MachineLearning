```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
X = np.array([1, 2, 3, 4])
y = np.array([20, 40, 50, 70])

# Inicialización
beta_0, beta_1 = 0, 10
alpha = 0.01
m = len(X)

# Listas para almacenar los valores del error y los parámetros
error_history = []
beta_0_history = []
beta_1_history = []

# Descenso de gradiente
for i in range(100):
    h = beta_0 + beta_1 * X
    error = h - y
    current_error = np.sum(error**2) / (2*m) #Error cuadrático medio
    error_history.append(current_error)
    beta_0_history.append(beta_0)
    beta_1_history.append(beta_1)

    #Gradiente descendente
    #beta_0 -= alpha * (1/m) * np.sum(error)
    beta_0 = beta_0 - alpha *  np.sum(error)/m
    #beta_1 -= alpha * (1/m) * np.sum(error * X)
    beta_1 = beta_1 - alpha * np.sum(error * X)/m



    # Visualización por iteración
    if (i + 1) % 1 == 0 or i == 0: # Muestra cada 100 iteraciones y la primera
        print(f"Iteración {i+1}: Error = {current_error:.2f}, beta_0 = {beta_0:.2f}, beta_1 = {beta_1:.2f}")

print(f"\nEcuación final: y = {beta_0:.2f} + {beta_1:.2f}x")

# Gráfica del error
plt.figure(figsize=(10, 6))
plt.plot(range(len(error_history)), error_history)
plt.xlabel("Iteration")
plt.ylabel("Mean Square Error")
plt.title("Error computation")
plt.grid(True)
plt.show()
```
